```python
# Live 4 (O) — Orquestração
# Notebook: 10_dim_catalogo_produtos_link
# Objetivo: não duplicar dimensão. Criar VIEW no tgt_schema apontando para a dimensão existente.

from pyspark.sql import functions as F

dbutils.widgets.text("src_schema", "aulas_ao_vivo.live_20260309")
dbutils.widgets.text("tgt_schema", "aulas_ao_vivo.live_20260323")

SRC_SCHEMA = dbutils.widgets.get("src_schema")
TGT_SCHEMA = dbutils.widgets.get("tgt_schema")

SRC_DIM = f"{SRC_SCHEMA}.silver_catalogo_produtos"
TGT_DIM = f"{TGT_SCHEMA}.silver_catalogo_produtos"  # view

spark.sql(f"CREATE SCHEMA IF NOT EXISTS {TGT_SCHEMA}")

spark.sql(f"""
CREATE OR REPLACE VIEW {TGT_DIM} AS
SELECT * FROM {SRC_DIM}
""")

df = spark.table(TGT_DIM)
print("Linhas (dim view):", df.count())
df.printSchema()
display(df.orderBy(F.col("produto").asc()).limit(50))
```
